```sos
from Bio import SeqIO
import os 
from amas import AMAS
```

```sos
all_genes = {}
#creates list of all genes 
for file in os.listdir("/Users/sarahharrison/Documents/GitHub/CF-BM-WGS-data-analysis/reference_data/muscle_test/gene_TEST"):
    if file.endswith(".fasta") == True:
        for seq_record in SeqIO.parse(file, "fasta"):
            gene_name = seq_record.id.split('_')[0]
            if gene_name not in all_genes:
                all_genes[gene_name] = []
                
print(all_genes)  
```

```sos
for file in os.listdir("/Users/sarahharrison/Documents/GitHub/CF-BM-WGS-data-analysis/reference_data/muscle_test/gene_TEST"):
    if file.endswith(".fasta") == True:
        for seq_record in SeqIO.parse(file, "fasta"):
            name = seq_record.id.split('_')[0]
            all_genes[name].append(seq_record)

print(all_genes)
for k,v in all_genes.items():
    SeqIO.write(v, k + "gene.fasta", "fasta") 
```
